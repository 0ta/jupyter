```python
dice_data = np.array([1, 2, 3, 4, 5, 6])

# サイコロを1000回振ってみる
calc_steps = 1000

# seedの固定
random.seed(0)

# 1〜6のデータの中から、1000回の抽出を実施
count_all_dice = random.choice(dice_data, calc_steps)

#print(count_all_dice == 1)
count1 = count_all_dice == 2
#print(count1)
n = sum(x == True for x in count1)
print(count_all_dice[count_all_dice == 1].size)
print(1000 // 6)
print(n)
```

```python
#一様分布
binlist = np.array([], dtype = 'float128')
binlist = random.rand(10000)
plt.hist(binlist)
plt.grid(True)
print(str(np.var(binlist)) + " 期待値:" + str(1/12))
print(str(np.mean(binlist)) + " 期待値:" + str(1/2))
```

```python
#一様分布->正規分布
random.seed(0)
sumval = 10000

errlist = np.array([], dtype = 'float128')
for _ in range(10000):
    count_all_bin = random.rand(sumval)
    #print(count_all_bin.mean())
    #print(count_all_bin.var())    
    errsumval = ((count_all_bin.mean() * sumval) -  sumval/2) / math.sqrt(sumval/12)
    #print(errval)
    errlist = np.append(errlist, errsumval)

plt.hist(errlist)
plt.grid(True)
```
